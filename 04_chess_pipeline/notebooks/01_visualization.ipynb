```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw_data/chess_games.csv")

df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df['winner'].value_counts().plot(
    kind='bar',
    title='Wins by Color',
    rot=0
)

plt.show()
```

```python
df['winner'].value_counts().plot(
    kind='bar',
    title='Wins by Color',
    color=['#C9A84C', '#1B3A2D', '#7A8C7E'],
    rot=0
)

plt.show()
```

```python
df.plot(
    kind='scatter',
    x='white_rating',
    y='turns',
    alpha=0.3,
    title='Rating vs Game Length',
    color='#3D6B4F'
)

plt.show()
```

```python
df[['white_rating', 'turns']].describe()
```

```python
df.boxplot(
    column='turns',
    by='victory_status',
    figsize=(8, 5)
)

plt.show()
```
