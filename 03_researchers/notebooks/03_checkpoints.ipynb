```python
import pandas as pd

df = pd.read_csv("../raw_data/researchers.csv")
df.head()
```

```python
df.info()
```

```python
df_filtered = df[(df['is_active'] == True) & (df['h_index'] > 15)]
print(f"Number of active researchers with h-index > 15: {len(df_filtered)}")
print(df_filtered[['first_name', 'last_name', 'h_index', 'is_active']])
```

```python
df_sorted = df_filtered.sort_values('joined_year', ascending=True)
```

```python
letters = df_sorted['last_name'].str[0]
print("".join(letters))
```

```python
df_funding = pd.read_excel("../raw_data/funding.xlsx")
```

```python
df_funding['amount_cad'] = pd.to_numeric(
    df_funding['amount_cad'],
    errors='coerce'
)
```

```python
clean_funding = df_funding[df_funding['amount_cad'] > 0]
```

```python
total = clean_funding['amount_cad'].sum()
print(total)
```
