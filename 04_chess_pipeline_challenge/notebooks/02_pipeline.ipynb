```python
import pandas as pd

# Load data
df = pd.read_csv("../data/raw/chess_games.csv")
```

```python
# ── 2a: Parse time_increment
df[['time_base', 'time_inc']] = (
    df['time_increment']
    .str.split('+', expand=True)
    .astype(int)
)
```

```python
# ── 2b: Add rating_diff
df['rating_diff'] = df['white_rating'] - df['black_rating']
```

```python
# ── 2c: Extract opening_family
df['opening_family'] = (
    df['opening_fullname']
    .str.split(':')
    .str[0]
    .str.strip()
)
```

```python
# ── 2d: Drop high-null column
df = df.drop(columns=['opening_response'])
```

```python
# ── 2e: Flag short/suspicious games ─────────────
df['is_suspicious'] = df['turns'] < 5
```

```python
# ── 2f: Validation checks ───────────────────────
assert df['rating_diff'].notna().all()
assert df.duplicated().sum() == 0
print("Stage 2 completed successfully")
```
