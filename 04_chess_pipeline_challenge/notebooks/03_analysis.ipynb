```python
import pandas as pd

df = pd.read_csv("../data/raw/chess_games.csv")
```

```python
# ─────────────────────────────
# Q10: Win rate (White / Black / Draw)
# ─────────────────────────────
q10 = df['winner'].value_counts(normalize=True) * 100
print("Q10 - Win Rates (%)\n", q10, "\n")
```

```python
# ─────────────────────────────
# Q11: Most common game end
# ─────────────────────────────
q11 = df['victory_status'].value_counts().idxmax()
print("Q11 - Most common end:", q11, "\n")
# Q11 - Most common end: Resign 
```

```python
# ─────────────────────────────
# Q12: Highest average turns by victory_status
# ─────────────────────────────
q12 = df.groupby('victory_status')['turns'].mean()
print("Q12 - Avg turns by status:\n", q12, "\n")
# Draw 83.781457
```

```python
# ─────────────────────────────
# Q13: Most popular opening_family when White/Black wins
# ─────────────────────────────
# Sicilian Defense
df['opening_family'] = df['opening_fullname'].str.split(':').str[0].str.strip()
```

```python
q13_white = df[df['winner'] == 'White']['opening_family'].value_counts().head(1)
```

```python
q13_black = df[df['winner'] == 'Black']['opening_family'].value_counts().head(1)
```

```python
print("Q13 - White openings:\n", q13_white)
```

```python
print("Q13 - Black openings:\n", q13_black, "\n")
```

```python
# ─────────────────────────────
# Q14: Rated vs Unrated White win rate
# ─────────────────────────────
q14 = df.groupby('rated')['winner'].apply(
    lambda x: (x == 'White').mean() * 100
)

print("Q14 - White win rate (rated vs unrated):\n", q14, "\n")
```

```python
# ─────────────────────────────
# Q15: Short / Medium / Long games
# ─────────────────────────────
def classify(turns):
    if turns < 20:
        return 'Short'
    elif turns < 60:
        return 'Medium'
    return 'Long'

df['game_length'] = df['turns'].apply(classify)

q15 = df['game_length'].value_counts(normalize=True) * 100

print("Q15 - Game length %:\n", q15)
```
